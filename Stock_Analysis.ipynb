# C-CITI-BANK STOCK PRICE HISTORY CHARTS

```python
citi=pd.read_csv('c.csv')
citi['Date']=citi['Date'].astype('datetime64[ns]')
citi.info()
```

# Bar plot showing total volume over the years

```python
citi["Year_c"]=citi['Date'].dt.year
yearly_volume=citi.groupby("Year_c")['Volume'].sum().reset_index()

sns.barplot(data=yearly_volume,x="Year_c",y="Volume")
plt.title('Total Trading Volumne by Year')
plt.xlabel('Year')
plt.xticks(rotation=45)
plt.show()
```

# Heatmap showing relationship between various apspects

```python
plt.figure(figsize=(10,8))

sns.heatmap(data=citi[['Open','Close','Low','High','Volume']].corr(),annot=True,cmap='coolwarm',center=0,linewidth=.5)
plt.title("Correlation Matrix of Stock Prices and Volume")
plt.show()
```

# GS- GOLDMANSACH STOCK PRICE HISTORY CHARTS

```python
gold=pd.read_csv('gs.csv')
gold['Date']=gold['Date'].astype('datetime64[ns]')
gold.info()
```

# BOX Plot showing daily closing balance from 2010 - 2019

```python
gold['Year_g']=gold['Date'].dt.year
filtered_gold=gold[(gold['Year_g']>=2010) & (gold['Year_g']<=2020)]


plt.figure(figsize=(14,7))
sns.boxplot(data=filtered_gold,x='Year_g',y='Close',palette='deep',hue='Year_g',dodge=False)
plt.title('Daily Closing Prices By Year (2010-2020)')
plt.xlabel("Year")
plt.xticks(rotation=45)
plt.show()
```

# Pair plot

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create the pair plot
pair_plot = sns.pairplot(data=gold)
pair_plot.fig.suptitle("Pair-Plot", y=1.02)  # Adjust title position

# Display the plot
plt.show()
```

# JPM- JP MORGAAN BANK STOCK PRICE HISTORY CHARTS

```python
jpm=pd.read_csv('JPM.csv')
jpm['Date']=jpm['Date'].astype('datetime64[ns]')
jpm.info()
```

# HISTOGRAM showing distribution of trading volume

```python
from matplotlib.ticker import FuncFormatter
plt.figure(figsize=(14,7))
sns.histplot(jpm['Volume'],bins=20,kde=True,color='green')

def millions(x,pos):
    return '%1.1fM' % (x * 1e-6)
plt.gca().xaxis.set_major_formatter(FuncFormatter(millions))
plt.title('Distribution of Trading Volume')
plt.show()
```

# MS-MORAGAN STANLEY STOCK PRICE HISTORY CHARTS

```python
morgan=pd.read_csv('MS.csv')
morgan['Date']=morgan['Date'].astype('datetime64[ns]')
morgan.info()
```

# Count Plot showing

```python

morgan['Price_Category'] = ['High' if x > morgan['Close'].mean() else 'Low' for x in morgan['Close']]

# Create the count plot
plt.figure(figsize=(14, 7))
sns.countplot(data=morgan, x='Price_Category', palette='coolwarm')
plt.title('Count of High and Low Prices')
plt.xlabel('Price Category')
plt.ylabel('Count')
plt.show()
```
